```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv('cities.csv')
```

```python
data
```

```python
data.head
```

```python
x=data['Area'].values
y=data['Pop 1980'].values
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
```

```python
m=len(x)
x=x.reshape((m,1))
```

```python
reg=LinearRegression()
reg=reg.fit(x,y)
```

```python
y_pred=reg.predict(x)
```

```python
plt.plot(x,y_pred)
```

```python
plt.plot(x,y_pred,'blue')
plt.scatter(x,y)
```
